```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace (0, 50, 1000)
y = 0.75 * np.sin(x)

peaks = np.where((y[1:-1] > y[0:-2]) * (y[1:-1] > y[2:]))[0] + 1
dips = np.where((y[1:-1] < y[0:-2]) * (y[1:-1] < y[2:]))[0] + 1

print(x[peaks])

plt.plot (x, y)
plt.plot (x[peaks], y[peaks], 'o')
plt.plot (x[dips], y[dips], 'o')

plt.show()
```

```python
# find zeroes
import numpy as np
import matplotlib.pyplot as plt

def findStatPoints(x,y):
    peaks = np.where((y[1:-1] > y[0:-2]) * (y[1:-1] > y[2:]))[0] + 1
    dips = np.where((y[1:-1] < y[0:-2]) * (y[1:-1] < y[2:]))[0] + 1
    
    print(x[peaks])
    print(y[peaks])
    
    return [[x[peaks], y[peaks]],[x[dips], y[dips]]]


x = np.array(range(0,5000,1))

y = np.sin(0.01*x)

points = findStatPoints(x,y)

#########
plt.plot(x,y)
plt.plot(points[0][0],points[0][1],'o')
plt.plot(points[1][0],points[1][1],'o')
#########


temp = [ x[points[0][0][0]:points[1][0][0]] ,y[points[0][0][0]:points[1][0][0]] ]
print(points[0][0][0]-1)

plt.plot(temp[0],temp[1])
```
